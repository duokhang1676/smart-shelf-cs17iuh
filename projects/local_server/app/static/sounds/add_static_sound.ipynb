```python
from gtts import gTTS
import vlc

def speech_text(text, path):
    tts = gTTS(text=text, lang='vi', slow=False)
    tts.save(path)
    player = vlc.MediaPlayer(path)
    player.play()

def play_sound(path):
    player = vlc.MediaPlayer(path)
    player.play()
```

```python
speech_text("Cảnh báo chưa thanh toán!","unpaid_warning.mp3")
```
